                                                                EDA

Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables

```python
match_ids_dict = {}

for index, row in match_resultsDF.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['scorecard']
    match_ids_dict[key2] = row['scorecard']
```

```python
#match_ids_dict
```

                                    BASIC TRANSFORMATION AND CLEANUP  IN BATTING DATA

```python
data_battingDF['out/not_out'] = data_battingDF.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
data_battingDF.head(11)
```

```python
data_battingDF.drop(columns=["dismissal"], inplace=True)
data_battingDF.head(11)
```

```python
data_battingDF['batsmanName'] = data_battingDF['batsmanName'].apply(lambda x: x.replace('â€', ''))
data_battingDF['batsmanName'] = data_battingDF['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
data_battingDF.head(30)
```

                            BASIC TRANSFORMATION AND CLEANUP  IN PLAYER DATA

```python
player_infoDF['name'] = player_infoDF['name'].apply(lambda x: x.replace('â€', ''))
player_infoDF['name'] = player_infoDF['name'].apply(lambda x: x.replace('†', ''))
player_infoDF['name'] = player_infoDF['name'].apply(lambda x: x.replace('\xa0', ''))
player_infoDF.head()
```

                                                CREATE FK COLUMN ON THE BASIS OF DICTNARY

```python
data_bowlingDF['scorecard'] = data_bowlingDF['match'].map(match_ids_dict)
data_bowlingDF.head()
```

```python


data_battingDF['scorecard'] = data_battingDF['match'].map(match_ids_dict)
data_battingDF.head()
```
